```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create plot and Change the line color to green.
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
ax.plot(x_axis, y_axis, marker="*", color="red", linewidth=2, label="Boston")
# Add a title and axes labels.
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title('Pyber Fare by Month')
# Add a y-axis limit.
ax.set_ylim(0, 45)
# Add grid lines.
ax.grid()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
# Create labels for the x and y axes.
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
# Create a title.
ax.set_title("PyBer Fare by Month")
```

```python
plt.plot(y_axis, x_axis, 'o', color='red', label='Chicago')
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
plt.xlim(0, 50)
# Add the legend.
plt.legend()
```
